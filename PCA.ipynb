Qn. Implement dimensionality reduction using **PCA**

1. Use the Iris Dataset present in the scikit-learn library
2. implement PCA to reduce dimension into 2 dimensions [PC1, PC2]
4. Display new reduced dimension values
5. Plot appropriate graph/ label w.r.t. principal component.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
```

```python
# Loading the dataset
iris = datasets.load_iris()

X = iris.data
y = iris.target
print("X:",X[0])
target_names = iris.target_names


# Scaling the data
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# Define a function for plotting graphs
def plot2clusters(X, y, target_names, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap='viridis', alpha=0.7)
    
    for i, name in enumerate(target_names):
        plt.scatter([], [], label=name, color=plt.cm.viridis(i / len(target_names)))

    plt.title(title)
    plt.xlabel('Principal Component 1 (PC1)')
    plt.ylabel('Principal Component 2 (PC2)')
    plt.legend()
    plt.colorbar(label='Target class')
    plt.show()


# Implement and visualize PCA
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)

print("Reduced dimensions: ", X_pca[:5])
plot2clusters(X_pca,y,target_names, 'PCA of Iris Dataset (2D Projection)')
```
